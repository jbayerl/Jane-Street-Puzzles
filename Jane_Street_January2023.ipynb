```python
'''Function printing the squares and the number of squares for any quadruple.'''
def number_of_squares(tuple):
    a = tuple[0]
    b = tuple[1]
    c = tuple[2]
    d = tuple[3]
    count = 1
    print(a,b,c,d)
    while a != 0 or b != 0 or c != 0 or d != 0:
        a_1 = abs(a - b)
        b_1 = abs(b - c)
        c_1 = abs(c - d)
        d_1 = abs(d - a)
        a,b,c,d = a_1,b_1,c_1,d_1
        print(a,b,c,d)
        count +=1
    return count
```

```python
def tribonacci(sequence):
    n = 0
    while n <= 10000000:
        n = sequence[-1] + sequence[-2] + sequence[-3]
        sequence.append(n)
    return sequence    
```
